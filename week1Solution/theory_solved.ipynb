## (Problem 1.12 from 'Learning By Data')

Given $y_1\leq\cdots\leq y_n\in \mathbb{R}$ find $h\in \mathbb{R}$ that on average is closest to $y_1,\ldots,y_n$ measured by squared distance (least squares). That is, 
$$
h_\textrm{ls} =\textrm{arg}\min_h \sum_{i=1}^n (h-y_i)^2
$$ 
<b>Question 1: </b>Show that
$h_\textrm{ls} = \frac{1}{n} \sum_{i=1}^n y_i$ is the minimizer. 

HINT: Computing the <a href="https://en.wikipedia.org/wiki/Chain_rule">derivative</a> may be worth the time and strain on your brain.

HINT: a local minimum is a global minimum!


<b>Question 2: </b>Consider absolute deviation instead of squared distance, i.e.

$$h_\mathrm{mad} =\textrm{arg}\min_h \sum_{i=1}^n |h-y_i|$$ 

Show tha $h_\mathrm{mad} = \mathrm{median}(y_1,\dots,y_n)$, the median of the $y$ values is the minimizer. 

HINT: Computing derivative may be usefull but  $|a|$ is not differentiable at zero but you may set it to zero (ask google about subgradients if you are interested).  

HINT: You can also argue purely algorithmically by thinking what happens with the cost as we sweep *h* from $-\infty$ to $\infty$).

HINT: a local minimum is a global minimum!

<b>Question 3: </b>What happens to the solutions $h_\mathrm{mean}, h_\mathrm{med}$ if we
add noise the last element $y_n$, i.e. $y_n = y_n + \varepsilon$ for
$\varepsilon \rightarrow \infty$.


### solution math
Least Squares:
$$
f(h) = \sum_{i=1}^n (h-y_i)^2, \quad f'(h) = \sum_{i=1}^n 2(h-y_i)
$$
set to zero
$$
\sum_{i=1}^n 2(h-y_i) = 0 \Rightarrow h = \frac{1}{n} \sum_{i=i}^n y_i
$$

Least Absolute Value:

You can do the same method except you need the gradient of abs which requires subgradients to be formal. For now say it is 0 at 0, -1 for negative x, and +1 for positive x.

Purely intuitive algorithhmic argument: Argue what happens with the cost as we move h from  negative infinity to positive infinity. 
The elements $y$ you move closer to reduce their cost in the sum linearly, i.e. if you move $h$ by $\delta$ than all of these decrease by $\delta$ (unless you pass one of them).
Similar for the points $y$ you move away from. If you are moving closer to more elements than you are moving away from you are decreasing cost, and you are increasing cost if you move away from more points than you move closer to.
Alas you must find a point that has an equal amount of elements that are smaller and larger...... The Median!!!

$$
f(h) = \sum_{i=1}^n |h-y_i|, \quad f'(h) = \sum_{h > y_i} 1 + \sum_{h < y_i} -1
$$
let $a = |\{h > y_i\|, b =|\{h < y_i\|$
For the gradient to be zero it must be the case that $a=b$ which indeed is a local minimum.
Thus you must find $h$ such that the number of $y$ larger than $h$ is equal to the number of $y$ less than $h$. The median.

In this exercise your job is to describe an algorithm that given a data set of labelled data (two classses only), constructs the binary classification tree (one internal node and two leafs) that minimize the 0-1 Loss over the training data. Such small classification trees are called classification stumps.

I.e. Given data 
$$ D = \{(x_i, y_i) \mid 1\leq i \leq n, y_i \in \{0, 1\}, x_i \in \mathbb{R}^d\}$$
construct the binary classification tree $T$ that minimize 
$$\frac{1}{n} \sum_{i=1}^n 1_{T(x_i) \neq y_i}$$.
Your algorithm must only use $O(n d \log n)$ time. 

The root node considers only questions like $f_i < 42$ and may this be represented by the features index i and the value to compare with (42 here).


### solution math
For a given feature we sort the feature and get the permutation and use that to sort the labels the same way.
Now for each value of the feature in this sorted order we need to compute the in sample cost cost of using this as root note.
We will scan the values from left to right and maintain the following values for the two leaves $l, r$ (left, right)
- $c_0^l$ the number of data points with zero labels that ends in the left leaf
- $c_1^l$ the number of data points with one label that ends in the left leaf
- $c_0^r$ the number of data points with zero labels that ends in  the right leaf
- $c_1^r$ the number of data points with one label that ends in the right leaf

If we know these values for a given root node (split) then we can compute the cost of that split simply as $\frac{1}{n} \cdot (\min(c_0^l, c_1^l) + \min(c_0^r, c_0^r))$
since the min of the values for a leaft is the number of mispredictions in the data from that leaf. Summing over the leafs gives the total number of mispredictions with the current split.
The normalizing $1/n$ is irrelevant for determining the best split!

Now we need to show how to maintain these as the algorithm progresses. Initially, all data is in one of the leafs and the other one is empty.
When we scan the different root nodes in sorted order, each step only changes these values by at most one depending on the data point we move from one leaf to the other and we can simply update the counts.

We have decided for you to present a Regression Stump by 
- idx: the data/feature vector index to consider in the root node (the one question asked)
- val: the value to compare to for data feature idx in the root node
- left: the value to return for a data point if it ends up in left leaf (x[idx] < val) (only question type we consider in a node)
- right: the value to return for a data point if it ends up in the right leaf (x[idx] >= val)

The approach we follow is as follow. Assume the input data has n data points each a vector of $d$ real numbers.

**Basic Algorithm:**

For each data feature f:
-   Compute for all possible values $v$ for feature $f$ in the training data the least squares cost of the stump achieved by using feature f and value $v$ in the root using the optimal value in the two leafs. This gives a list of of costs, one for each split ($f, v: \textrm{cost}$). 
- Pick the split $f, v$ with minimal cost and create the corresponding tree by setting idx, val, left, right
